**Label Encoder Method**


- LabelEncoder method is used convert catagoricsl data to numerical data

- Labelencoder method is same as for loop in map

- It is under  **sklearn** package i.e **scikit -learn**

- under sklearn we have a class called as preprocessing 

- Because we are doing data processing only

- under data processing we a method called as **LabelEncoder**

- sklearn
    - preprocessing
        -  LabelEncoder


**3 steps in sklearn package**

1.  Read the package

2.  Save the package

3.  Apply the fit transform on data

```python
#1. Read the package\ load the method
#from sklearn(file/package).preprocessing(class) import LabelEncoder(Method name)

from sklearn.preprocessing import LabelEncoder

#2.  Save the package/call the method

le=LabelEncoder()

#3.  Apply the fit transform on data

telecom_df['mail_subscribed']=le.fit_transform (telecom_df['mail_subscribed'])
telecom_df

```

```python
from sklearn.preprocessing import LabelEncoder

telecom_df=pd.read_csv('telecom_churn_data.csv')
cat_column=telecom_df.select_dtypes(include='object').columns

le=LabelEncoder()

for i in cat_column:
     telecom_df[i]=le.fit_transform (telecom_df[i])
telecom_df     

#All catogorical column converted into numerical column
```

**Fit and Transform**

- Fit means develop the application and 

- transform means process the application

- in map method we develop dictionary logic and then we apply that dictionary to colum

- develop dictionary means applying fit 

- apply dictionary to column means transforming 

- whenevar we transforming or changing data then use **fit-transformation**

- if we are not changing the data then only fit.


**np.where**

- Using np.where also we can also used to change catagorical to numerical column

- but only used for binary condition i.e., like if and else

- if condition is true: will apply value

- if condition is false: will apply false value

- so np.where used in binary label

- so column should have only two unique labels

```python
from sklearn.preprocessing import LabelEncoder

telecom_df=pd.read_csv('telecom_churn_data.csv')

con=telecom_df['mail_subscribed']=='no'

telecom_df['mail_subscribed']=np.where(con,0,1)
telecom_df
```

```python
from sklearn.preprocessing import LabelEncoder

telecom_df=pd.read_csv('telecom_churn_data.csv')

con=telecom_df['gender']=='Male'

telecom_df['gender']=np.where(con,0,1)
telecom_df
```
