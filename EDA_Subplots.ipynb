**Subplots**

- Plots multiple graphs

- suppose 2*2 matrix and 4 plots are there

- then 1st possition - (2,2,1)

- 2nd (2,2,2)

- 3rd (2,2,3)

- 4rth (2,2,4)

```python
import matplotlib.pyplot as plt
```

```python
plt.subplot(2,2,1)
plt.subplot(2,2,2)
plt.subplot(2,2,3)
plt.subplot(2,2,4)
```

```python
for i in range(1,5):
    plt.subplot(2,2,i)
```

```python
for i in range(1,7):
    plt.subplot(3,3,i)
```

```python
plt.subplot(1,2,1)
plt.subplot(1,2,2)
```

```python
plt.subplot(2,1,1)
plt.subplot(2,1,2)
```
